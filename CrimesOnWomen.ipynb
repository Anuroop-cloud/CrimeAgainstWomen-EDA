```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
desc = pd.read_csv("description.csv")
desc
```

```python
df = pd.read_csv("CrimesOnWomenData.csv")

df.head()
df.info()
df.describe()  #numerical data distribution
```

```python
#dropping duplciates
df = df.drop_duplicates()

#handling missing vals
df = df.fillna(0)

df["State"] = df["State"].str.lower().str.strip()   #realised case insensitivity
#grouping state crimes
state_crime = df.groupby("State").sum(numeric_only=True)
```

###Q1: States with highest crime aganist women

```python
state_crime["Total_Crime"] = state_crime.sum(axis=1)
top_states = state_crime["Total_Crime"].sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_states.values, y=top_states.index)
plt.title("Top 10 States with Highest Crime Against Women")
plt.xlabel("Total Crime Cases")
plt.ylabel("State")
plt.show()
```

###Q2: Clustering States Based on Crime Data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X = state_crime.drop(columns=["Total_Crime"]) #to avoid dominance over individual crime patterns

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
state_crime["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
sns.scatterplot(
    x=state_crime.iloc[:,0],
    y=state_crime.iloc[:,1],
    hue=state_crime["Cluster"],
    palette="Set1"
)
plt.title("Clustering of States Based on Crime Data")
plt.show()
```
